```python
# libs necessárias: 

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import warnings

warnings.filterwarnings('ignore')
```

```python
Base_dados = pd.read_excel('/content/Vase_004 - Magalu - Sem Resolução.xlsx')

Base_dados.head()
```

```python
# Verificando a dimensão:

Base_dados.shape
```

```python
Base_dados.info()
```

```python
Base_dados.describe()
```

```python
# Series temporais: 

Dados = Base_dados.set_index("Data")

Dados.head()
```

```python
# Gráfico de fechamento:

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,5))
plt.title("Análise das ações da Magalu - Fechamento",fontsize=15,loc='left')
plt.plot(Dados.index,Dados['Fechamento']);
plt.xlabel("Período da Cotação")
plt.ylabel("Valor da Ação (R$)")
```

```python
# Detalhes dos dados:
Dados.tail()
```

```python
# Agrupando em 5 em 5 dias e tirando a média movel.
Media_Movel = Dados['Fechamento'].rolling(5).mean()
# Agrupando em 30 em 30 dias e tirando a média movel.
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,5))
plt.title("Análise das ações da Magalu - Fechamento",fontsize=15,loc='left')

plt.plot(Dados.index,Dados['Fechamento']);
plt.plot(Dados.index,Media_Movel);
plt.plot(Dados.index,Media_Tendencia);


plt.xlabel("Período da Cotação")
plt.ylabel("Valor da Ação (R$)")
```

```python
sns.boxplot(data = Dados,x='Fechamento');
```

```python
# Boxplot mensal

Base_dados['Mes'] = Base_dados['Data'].dt.month

plt.figure(figsize=(16,5))
sns.boxplot(data=Base_dados,x='Mes',y='Fechamento');
```

```python
Base_dados.groupby(['Mes']).describe()['Fechamento']
```

```python
# Grafico

Grafico = go.Figure(
    data=[
        go.Candlestick(
            x = Dados.index,
            open = Dados['Abertura'],
            high = Dados['Maior'],
            low = Dados['Menor'],
            close = Dados['Fechamento'],
        )
    ]
)
Grafico.update_layout(xaxis_rangeslider_visible=False)
Grafico.show()
```
